# Let's crawl all of them

針對每一個連結項目抓出標題、連結、日期等資訊

```python
import requests
from bs4 import BeautifulSoup
import re
from datetime import datetime
import time
```

```python

category_url = 'https://www.sportsv.net/baseball/mlb'

# req = requests.get(category_url)
req = requests.get(category_url, headers={ 'user-agent': user_agent.random })

# page = BeautifulSoup(req.text, 'lxml')
page = BeautifulSoup(req.text, 'html.parser')
```

```python
req.text
```

```python
# 抓新聞列表
items = newpage.findAll('div',{'class':"row item"})
items
```

```python
category = 'MLB'


# 存放資料之變數
article_link_list =[]
title_list=[]
date_list=[]
category_list=[]
photo_link_list=[]

item_id_list=[]
content_list=[]


serial_no = 1

for item in items:

    #新聞標題
    title = item.find('h3').text.strip()
    cont = re.sub(r'.*(東京賽》|MLB》)', '', title)
    print( cont )

    #新聞日期
    news_time = item.find('div', {'class':"date"}).text
    print(news_time)
    dtime = datetime.strptime(news_time, '%Y/%m/%d')
    news_date = dtime.strftime("%Y-%m-%d")
    
    #新聞連結
    article_url = item.find('a').get('href')
    print(article_url)

    # 最新的新聞列表之新聞連結:開頭缺少"https://www.cna.com.tw"
    # if article_url.startswith("/news"):
    #     article_url = "https://www.cna.com.tw"+article_url
    #     article_link_list.append(article_url)
    # else:
    #     article_link_list.append(article_url)
    
    


    try:
    # Use if else to deal with it
        if item.find('img').get('src'):
            photo_link = item.find('img').get('src')
    except:
        print('No photo link!')
        photo_link=''
    
    print(photo_link)

    
    category_list.append(category)
    date_list.append(news_date)
    title_list.append(cont)
    photo_link_list.append(photo_link)
    article_link_list.append(article_url)
    
    # item id
    date_str = dtime.strftime("%Y%m%d")

    item_id_list.append("sportsv_" + date_str + "_" + str(serial_no))
    serial_no += 1

    ##### Crawl article content爬新聞內容
    # There are some words we don't like to analyze, They should be removed.
    req = requests.get(article_url, headers={ 'user-agent': user_agent.random }, timeout=5)
    
    # page = BeautifulSoup(req.text,'lxml')
    page = BeautifulSoup(req.text, 'html.parser')

    # Find content
    content_div = page.find('div', class_='article-content')
    paragraphs = content_div.find_all('p')
    content = ''.join([p.get_text(strip=True) for p in paragraphs])
    content = re.sub(r'1.大谷翔平.*', '', content)
    content


    # Remove unnecessary content
    # Remove "透過 Google News追蹤中央社"
    # try:
    #     content.find('div', {'class':"gmailNews"}).decompose()
    # except:
    #     pass

    # # Remove "訂閱早安電子報"
    # try:
    #     content.find('div', {'class':"subscriInnTitle"}).decompose()
    # except:
    #     pass

    # # 額外插入的訂閱連結-->移除之
    # try:
    #     content.find('div', {'class':"SubscriptionInner mySubscriptionInner"}).decompose()
    # except:
    #     pass

    # # 額外插入的延伸閱讀區塊-->移除之
    # try:
    #     content.find('div', {'class':"paragraph moreArticle"}).decompose()
    # except:
    #     pass
    # # 額外插入的圖片與文字說明區塊-->移除之
    # try:
    #     for media in content.findAll('div', {'class':"media"}):
    #         media.decompose()
    # except:
    #     pass

    # Remove unnecessary content
    # cont = content.text
    # cont = re.sub('（中央社.+?）','', cont)
    # cont = re.sub('（編輯：.*','', cont) 
    # cont = re.sub('（譯者：.*','', cont) 
    
    content_list.append(content)
    
    # scrap only 10 articles
    if serial_no >= 10:
        break
    
    time.sleep(10)  # 遵守爬蟲禮節，請小睡一下
```

```python
category_url = 'https://www.sportsv.net/baseball/cpbl'

# req = requests.get(category_url)
req = requests.get(category_url, headers={ 'user-agent': user_agent.random })

# page = BeautifulSoup(req.text, 'lxml')
page = BeautifulSoup(req.text, 'html.parser')
```

```python
req.text
```

```python
cpbl_items = page.findAll('div',{'class':"row item"})
cpbl_items
```

```python
item = cpbl_items[1]
```

```python
item
```

```python
item.find('h4').text.strip()
```

```python
category = '中職'

serial_no = 1

# 存放資料之變數
article_link_list = []
title_list = []
date_list = []
category_list = []
photo_link_list = []

item_id_list = []
content_list = []

for item in cpbl_items:
    # 新聞標題
    title = (item.find('h3') or item.find('h4')).text.strip() if (item.find('h3') or item.find('h4')) else None
    print(title)

    # 如果標題為 None，則跳過這篇新聞
    if title is None or title == "":
        print("標題為空，跳過這篇新聞。")
        continue

    # 新聞日期
    news_time = item.find('div', {'class': "date"}).text.strip() if item.find('div', {'class': "date"}) else ""
    print(news_time)

    # 新聞連結
    article_url = item.find('a').get('href')
    print(article_url)

    try:
        # Use if else to deal with it
        if item.find('img').get('src'):
            photo_link = item.find('img').get('src')
    except:
        print('No photo link!')
        photo_link = ''

    print(photo_link)

    category_list.append(category)
    date_list.append(news_time)
    title_list.append(title)  # 修正了這裡的變數名稱
    photo_link_list.append(photo_link)
    article_link_list.append(article_url)

    # item id
    date_str = dtime.strftime("%Y%m%d")

    item_id_list.append("sportsv_" + date_str + "_" + str(serial_no))
    serial_no += 1

    ##### Crawl article content 爬新聞內容
    # There are some words we don't like to analyze, They should be removed.
    req = requests.get(article_url, headers={'user-agent': user_agent.random}, timeout=5)
    # page = BeautifulSoup(req.text, 'lxml')
    page = BeautifulSoup(req.text, 'html.parser')
    


    # Find content
    content_div = page.find('div', class_='article-content')
    paragraphs = content_div.find_all('p')
    content = ''.join([p.get_text(strip=True) for p in paragraphs])
    content

    content_list.append(content)

    # scrap only 10 articles
    if serial_no >= 10:
        break

    time.sleep(10)  # 遵守爬蟲禮節，請小睡一下
```

```python
data = zip(item_id_list,date_list, title_list,category_list,content_list,article_link_list,photo_link_list)
df = pd.DataFrame(list(data), columns=['item_id','date','title','category','content','link','photo_link'])
```

```python
df.to_csv("sportsv_baseball_categroy.csv", sep="|", index=False, mode='a', header=False)
```
